```python
import control as ctrl
import matplotlib.pyplot as plt
import numpy as np

L = 1.12e-6;
R = 2.8;
K_t = 0.22;
K_e = 2.25;
B = 0.123;
K_b = 0.0;
J = 0.1;
tau_L = 0;

s = ctrl.TransferFunction.s

s = ctrl.TransferFunction.s
H = K_t/((K_t*K_e) + (s*L + R) * (J*s + B));
H
```

```python
t,y = ctrl.step_response(H);
plt.plot(t,y)
```

```python
omega = np.logspace(-1, 2, 500)
mag, phase, omega_out = ctrl.frequency_response(H, omega)

# freqresp/frequency_response returns arrays with shape (1,1,N), squeeze them
mag = mag.squeeze()
phase = phase.squeeze()

# Plot Bode manually
plt.figure(figsize=(8,6))

# Magnitude
plt.subplot(2,1,1)
plt.semilogx(omega, 20*np.log10(mag))
plt.ylabel('Magnitude (dB)')
plt.grid(True)

# Phase
plt.subplot(2,1,2)
plt.semilogx(omega, np.degrees(phase))
plt.ylabel('Phase (deg)')
plt.xlabel('Frequency (rad/s)')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# PID constants
K_p = 10000;
T_i = 0.0004;
T_d = 0.001;

C = K_p*(1 + (1/(T_i*s)) + (T_d*s) )
#C = K_p*(1 + (1/(T_i*s)));
#C = K_p*(1 + (T_d*s) );
#C = K_p;

T = (C*H)/(1 + C*H);

t_t,y_t = ctrl.step_response(T);
plt.plot(t_t,y_t)
plt.grid(True)
```
